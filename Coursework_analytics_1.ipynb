```python
#импортируем все, что можно
import pandas as pd
import numpy as np
from datetime import date
import matplotlib.pylab as plt
import seaborn as sns
from matplotlib.pylab import rcParams
import psycopg2
import pymongo
import missingno as msno
from datetime import timedelta, datetime
import matplotlib.pyplot
```

```python
#Загрузим каждый файл в соответствующий датасет
courses = pd.read_csv("courses.csv", index_col=0)
students = pd.read_csv("students.csv")
# удаляем из датасета столбец id, которого нет по заданию 
students = students.drop(students.columns[0], axis='columns')
course_c = pd.read_csv('course_contents.csv')
prog = pd.read_csv('progresses.csv')
prog_p = pd.read_csv('progress_phases.csv')
```

```python
#переименуем столбец идентификатора курса в courses так, как он называется в course_contents для последующего объединения таблиц
courses.rename({'id': 'course_id'}, axis=1, inplace=True)
#объединение courses и course_contents по столбцу course_id во временный датасет
temp1=courses.merge(course_c, on='course_id', how='left')
```

```python
#переименуем столбец идентификатора прогресса в progresses так, как он называется в progress_phases для последующего объединения таблиц
prog.rename({'id': 'progress_id'}, axis=1, inplace=True)
#объединяем prog b prog_p по столбцу progress_id во временный датасет 
temp2=prog.merge(prog_p, on='progress_id', how='left')
```

```python
# объединяем все в глабный датасет
main = pd.merge(temp1, temp2, on = ['module_number', 'lesson_number', 'course_id', ], how ='left')
main
```

```python
# переименовываем столбец в students для того, чтобы объединить таблицы
students.columns = ['student_id', 'city', 'birthday']
# создаем датасет для расчета для расчета среднего возраста студентов курсов
ages = pd.merge(main, students, on = 'student_id', how ='left')
ages.head()
```

```python
# посмотрим что получилось и много ли пропущенных значений в датасете
msno.bar(ages)
# пропущенных значений не так уж много. Больше всего их в столбце "Город", потом в - "Дата Рождения". Пропущенные значения в 
# в столбце "Дата окончания" могут говорить о том, что на момент формирования отчета не все студенты еще окончили курсы (или
# кто-то бросил учебу).
```

```python
# общее количество курсов в датасете
main['title'].nunique()
```

```python
# количество модулей на каждом курсе
main.groupby('title').module_title.nunique()
```

```python
#количество уроков в каждом модуле на каждом курсе
main.groupby(['title', 'module_title']).lesson_title.nunique()
```

```python
# медианное количество уроков в модуле на каждом курсе
main.groupby(['title', 'module_title']).lesson_title.nunique().groupby(['title']).median()
```

```python
#количество учеников на каждом курсе
main.groupby('title').student_id.nunique()
```

```python
#для решения следующих задач почистим данные и уберем пустые значения
ages.dropna(subset=['birthday'], inplace = True)
# посмотрим что получилось
msno.bar(ages)
#отлично. все пропущенные даты рождения удалены. Можно считать возраст тех студентов, по которым есть данные.
```

```python
# преобразуем даты
ages['start_date']=pd.to_datetime(ages['start_date'],utc=True,  yearfirst=True, errors='coerce',format='%Y-%m-%d')
ages['finish_date']=pd.to_datetime(ages['finish_date'],utc=True,  yearfirst=True, errors='coerce',format='%Y-%m-%d')
ages['birthday']=pd.to_datetime(ages['birthday'],utc=True,  yearfirst=True, errors='coerce',format='%Y-%m-%d')
```

```python
# добавим в датасет ages столбец age,
# возраст студентов считаем на момент начала обучения на курсе: start_date
ages['age'] = round((ages['start_date'] - ages['birthday']).apply(lambda x: x/np.timedelta64(1,'Y')), 0)
```

```python
# минимальный, максимальный, средний, медианный возраст студентов
ages['age'].agg(['min', 'max', 'mean', 'median'])
```

```python
# явно где-то ошибки в указании даты рождения студентов.
# отфильтруем только тех, чей возраст попадает в рамки от 14 до 80 лет.
ages = ages[(ages['age'] >= 14.0) & (ages['age'] <= 80.0)]
```

```python
# посмотрим заново минимальный, максимальный, средний, медианный возраст студентов
ages['age'].agg(['min', 'max', 'mean', 'median'])
# вот теперь адекватные данные
```

```python
# минимальный, максимальный, средний, медианный возраст студентов на каждом курсе
ages.groupby('title')['age'].agg(['min', 'max', 'mean', 'median'])
```

```python
# Постройте bar-chart, отражающий количество студентов на каждом курсе. Ticks нужно развернуть так, чтобы они были читаемы

# Определение основных объектов диаграммы
fig, ax = plt.subplots()

# Построение диаграммы, отражающей количество студентов на каждом курсе
main.pivot_table(
    values='student_id', index='title', aggfunc=lambda x: x.unique().size
).sort_values(
    'student_id', ascending=True
).plot.bar(
    ax=ax, legend=False
)

# Вывод подписей делений оси абсцисс
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')

# Вывод названия осей
ax.set(xlabel='Наименование курса', ylabel='Количество студентов')

# Вывод диаграммы на страницу
plt.show()
```

```python
# Постройте горизонтальный (столбцы должны располагаться горизонтально) bar-chart, отражающий количество студентов на каждом 
# курсе. График должен иметь заголовок. Значения должны быть отсортированы. Цвет столбцов должен содержать информацию о сфере,
# к которой относится курс (то есть нужна легенда). Прозрачность должна стоять на отметке 0.1. На график должна быть нанесена 
# линия медианы. У медианы должен быть свой цвет. Рамки у графика быть не должно

# Определение основных объектов диаграммы
fig, ax = plt.subplots()

# Формирование датасета по количеству студентов на курсах
# в разрезе сфер, к которому относится курс
chart = main.pivot_table(
    values='student_id', index='title', columns='field', aggfunc=lambda x: x.unique().size, margins=True
).sort_values(
    'All', ascending=True
).drop(
    index=['All'], columns=['All']
)

# Построение диаграммы, отражающей количество студентов на курсах
chart.plot.barh(ax=ax, title='Количество студентов на курсах', stacked=True, alpha=0.1)

# Построение линии медианы на диаграмме
chart = chart.stack().reset_index()
ax.vlines(chart[0].median(), ax.get_ylim()[0], ax.get_ylim()[1], color='r', linestyles='dashed', linewidth=1)

# Скрытие контура диаграммы
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
ax.spines['left'].set_color('none')
ax.spines['bottom'].set_color('none')

# Вывод названия осей
ax.set(xlabel='Количество студентов', ylabel='Наименование курса')

# Вывод диаграммы на страницу
plt.show()
```

```python
# Вариант 2

#Сделаем таблицу, где найдем для каждого курса номер модуля, в котором появляется первая домашняя работа
number = main[main['is_homework']==True][['title','module_number']].groupby('title')['module_number'].agg('min').reset_index()

# Сделаем датасет для оценки прироста студентов на курсах
# Сделаем датасет по первому курсу из таблицы выше
start = main[(main['title'] == number['title'][0]) & (main['module_number'] == number['module_number'][0])].groupby(['title', 'student_id'])['start_date'].min().reset_index()

# Формируем остальной датасет
for i in range(1, len(number['title'])):
    temp = main[(main['title'] == number['title'][i]) & (main['module_number'] == number['module_number'][i])].groupby(['title', 'student_id'])['start_date'].min().reset_index()
    start = pd.concat([start, temp], ignore_index=True)

# Преобразуем даты к краткому виду
start['start_date'] = start['start_date'].astype('datetime64[M]')

# Создание списка временного ряда за весь период сбора статистики в формате "Год-Месяц"
column_list = pd.date_range(
    start['start_date'].min(), start['start_date'].max() + pd.tseries.offsets.MonthEnd(0), freq='M'
).strftime('%Y-%m-01').to_list()

# Добавление временного ряда к датасету в качестве столбцов и проставление признака '1'
# в случае соответствия даты начала курса с датой (столбцом) из временного ряда
for item in column_list:
    start.loc[start['start_date']==item, item] = 1
    
# Преобразование меток столбцов и значений в строки
start = start.groupby(['title', 'student_id']).sum().stack().reset_index().rename(
    columns={'level_2': 'start_date', 0: 'start_date_tag'})

# Задание типа столбца
start['start_date'] = start['start_date'].astype('datetime64[D]')

# Формирование датасета по приросту студентов в разрезе каждого месяца
start = start.groupby(
    ['title', 'start_date']
).aggregate(
    student_count = ('start_date_tag', np.sum)
).reset_index()

# Расчет прироста студентов на каждом курсе в каждом месяце
for item in start['title'].unique():
    indexes = start.loc[start['title']==item, 'student_count'].keys()
    for index in indexes:
        start.loc[index, 'student_increment'] = start.loc[indexes[0]:index, 'student_count'].sum()
```

```python
# Рассчитать минимальное, максимальное, среднее, медианное время прохождения каждого модуля (разность между временем начала и 
# окончания выполнения домашней работы) для каждого курса. Если домашних заданий в модуле несколько, то считать разность между 
# временем начала выполнения первой домашней работы и временем окончания выполнения последней домашней работы в модуле.

# Задание типа для данных в столбцах 'start_date' и 'finish_date'
main['start_date'] = main['start_date'].astype('datetime64[ms]')
main['finish_date'] = main['finish_date'].astype('datetime64[ms]')

# Формирование датасета в разрезе модулей и курсов
modules = main.loc[(main['is_homework']==True) & (main['finish_date'].notnull()),
                         ['title', 'module_title', 'start_date', 'finish_date']]

# Форматирование дат
modules['start_date']=pd.to_datetime(modules['start_date'],utc=True,  yearfirst=True, errors='coerce',format='%Y-%m-%d')
modules['finish_date']=pd.to_datetime(modules['finish_date'],utc=True,  yearfirst=True, errors='coerce',format='%Y-%m-%d')

# Определение времени прохождения каждого модуля для каждого курса
modules['duration'] = (modules['finish_date'] - modules['start_date']).dt.days

# Очистка датасета от "аномальных" данных
modules = modules.query('finish_date >= start_date')


#Вывод статистики по прохождении каждого модуля и курса
modules.groupby(['title', 'module_title']).aggregate(
    {'start_date': np.min, 'finish_date': np.max, 'duration': [np.min, np.max, np.mean, np.median]}
).reset_index().sort_values(
    ['title', ('start_date', 'amin')]
).head()
```

3. **(m)** Постройте горизонтальный (столбцы должны располагаться горизонтально) bar-chart, отражающий конверсию перехода студентов из одного модуля в другой на каждом курсе. 15 графиков. Графики должны иметь заголовки. Ticks должны содержать номер и название модуля. Цвет столбцов графиков должен содержать информацию о сфере, к которой относится курс (нужна легенда). Прозрачность должна стоять на отметке 0.1. На графики должна быть нанесена линия медианы конверсии для каждого курса. У медианы должен быть свой цвет. Рамок у графиков быть не должно ⭐
4. На основании рассчитанных значений сделайте аналитический вывод (должен быть полным и покрывать все полученные выше метрики)

```python
# Посчитать конверсию перехода студентов из одного модуля в другой на каждом курсе. Формула: отношение количества студентов, 
# приступивших к выполнению домашнего задания в этом модуле (если дз в модуле несколько, то считать по первому дз в модуле), 
# к количеству студентов, сдавших задание в предыдущем модуле (если дз в модуле несколько, то считать по последнему дз в модуле).

# Формирование датасета через группировку по модулю и студенту
# с определением дат начала и окончания всех домашних заданий в модуле
conversion = main.loc[
    main['is_homework']==True, 
    ['title', 'field', 'module_number', 'module_title', 'student_id', 'start_date', 'finish_date']
].groupby(
    ['title', 'field', 'module_number', 'module_title', 'student_id']
).aggregate(
    {'start_date': np.min, 'finish_date': np.max}
)

# Перегруппировка датасета по модулю c подсчетом
# количества студентов начавших и завершивших обучение 
conversion = conversion.assign(
    student_start=conversion['start_date'].notnull(), student_finish=conversion['finish_date'].notnull()
).groupby(
    ['title', 'field', 'module_number', 'module_title']
).aggregate(
    {'student_start': np.sum, 'student_finish': np.sum, 'start_date': np.min, 'finish_date': np.max}
).sort_index().reset_index()

# Преобразование формата дат к краткому виду
conversion['start_date'] = conversion['start_date'].astype('datetime64[D]')
conversion['finish_date'] = conversion['finish_date'].astype('datetime64[D]')

# Расчет конверсии и добавление значений в датасет
for course in conversion['title'].unique():
    for index, module_number in conversion.loc[conversion['title']==course, 'module_number'].items():
        # Определение количества студентов, начавших обучение в текущем модуле
        student_start = conversion.loc[index, 'student_start']
        # Определение количества студентов, завершивших обучение в предыдущем модуле
        student_finish = conversion.loc[
            (conversion['title']==course) & (conversion['module_number'] < module_number), 'student_finish'
        ].tail(1).min()
        # Подсчет конверсии текущего модуля
        conversion.loc[index, 'conversion'] = student_start / student_finish if student_finish > 0 else 0

# Вывод первых 5-ти позиций
conversion.head()
```

```python
# Постройте bar-chart, отражающий конверсию перехода студентов из одного модуля в другой на каждом курсе. График должен иметь 
# заголовок. Ticks нужно развернуть так, чтобы они были читаемы

# Определение основных объектов диаграммы
fig, ax = plt.subplots(figsize=(26, 15))

# Вывод диаграммы на страницу
conversion.pivot_table(
    values='conversion', index=conversion.index, columns='title'
).plot.bar(
    ax=ax, title='Конверсия перехода студентов из одного модуля в другой на курсах', stacked=True
)

# Вывод подписей делений оси абсцисс
ax.set_xticklabels(conversion['module_title'].apply(lambda x: f'{x[:30]}...'), rotation=45, ha='right')

# Вывод подписей осей координат
ax.set_xlabel('Модули курса')
ax.set_ylabel('Конверсия')

# Автоматическое позиционирование элементов диаграммы
fig.tight_layout()

# Отображение диаграммы на странице
plt.show()
```

```python
# Постройте горизонтальный (столбцы должны располагаться горизонтально) bar-chart, отражающий конверсию перехода студентов 
# из одного модуля в другой на каждом курсе. 15 графиков. Графики должны иметь заголовки. Ticks должны содержать номер и 
# название модуля. Цвет столбцов графиков должен содержать информацию о сфере, к которой относится курс (нужна легенда). 
# Прозрачность должна стоять на отметке 0.1. На графики должна быть нанесена линия медианы конверсии для каждого курса. У 
# медианы должен быть свой цвет. Рамок у графиков быть не должно

# Импортирование библиотеки цветов matplotlib
import matplotlib.colors as colors

# Определение основных объектов диаграммы
fig, ax = plt.subplots(15, 1, figsize=(10, 50))

# Формирование списка курсов
courses = np.sort(conversion['title'].unique())

# Формирование словаря из сферы курса и цвета
colors = dict(zip(conversion['field'].unique(), colors.XKCD_COLORS))

# Вывод диаграмм на страницу
for index, ax in enumerate(ax.flat):
    # Определение датасета курса
    chart = conversion[conversion['title']==courses[index]].sort_values('conversion').reset_index(drop=True)
    # Вывод диаграммы на страницу
    chart.pivot_table(
        values='conversion', index=chart.index, columns='field'
    ).plot.barh(
        ax=ax, width=0.7, title=titles[index], color=colors[chart['field'].unique()[0]], alpha=0.1
    )
    # Вывод линии медианы на диаграмму
    ax.vlines(chart['conversion'].median(), ax.get_ylim()[0], ax.get_ylim()[1], color='r', linestyles='dashed', linewidth=1)
    # Вывод подписей делений оси ординат
    ax.set_yticklabels(
        chart[['module_number', 'module_title']].apply(
            lambda x: '{}. {}'.format(
                x['module_number'], 
                x['module_title'][:40] + '...' if len(x['module_title']) > 40 else x['module_title']), 
            axis=1))
    # Вывод подписей осей координат
    ax.set_xlabel('Конверсия перехода')
    ax.set_ylabel('Модули курса')
    # Скрытие рамки диаграммы
    ax.spines['top'].set_color(None)
    ax.spines['right'].set_color(None)

# Автоматическое позиционирование элементов диаграммы
fig.tight_layout()

# Отображение диаграмм на странице
plt.show()
```
